```python
import pandas as pd
import numpy as np
import sklearn
```

```python
from sklearn.datasets import load_boston
```

```python
boston = load_boston()
```

```python
boston.feature_names
```

```python
boston_df = pd.DataFrame(boston.data, columns=boston.feature_names)
```

```python
boston_df.head()
```

```python
boston_df.describe()
```

# Exporatory Data Analysis (EDA)

```python
boston_df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.pairplot(boston_df, height=1.5);
plt.show()
```

```python
cols_study = ['CRIM', 'ZN', 'INDUS', 'NOX', 'RM']
```

```python
sns.pairplot(boston_df[cols_study], height=2.5)
plt.show()
```

# Correlation Analysis and Feature selection

```python
pd.options.display.float_format = "{:,.3f}".format
```

```python
boston_df.corr()
```

```python
plt.figure(figsize=(16,10))
sns.heatmap(boston_df.corr(), annot=True)
plt.show()
```

```python
cols_to_study = ['CRIM', 'ZN', 'INDUS','CHAS']
```

```python
sns.heatmap(boston_df[cols_to_study].corr(), annot=True)
plt.show()
```
